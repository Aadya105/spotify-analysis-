```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#load the dataset
df=pd.read_csv('tracks.csv')
```

```python
# first ten rows of the dataset
df.head(10)
```

```python
#null values
df.isnull().sum()
```

```python
#handling missing values in the dataframe
df = df.dropna(subset=['name'])
print(pd.isnull(df).sum())
```

```python
df.info()
```

```python
#convert duration from ms to seconds
df["duration"]= df["duration_ms"].apply(lambda x: round(x/1000))
df.drop("duration_ms", inplace=True, axis=1)
```

```python
#sorting the dataset in ascending order of popularity
df=df.sort_values('popularity',ascending = True)
df.head(5)
```

```python
df.describe()
```

```python
# 5 most popular songs in the dataset
popular = df.query('popularity>80', inplace = False).sort_values('popularity', ascending= False)
popular[:5]
```

```python
#5 least popular songs
least_popular = df[['name','artists','popularity']].sort_values('popularity', ascending = True )
least_popular[:5]
```

```python
sample_df = df.sample(int(0.005*len(df)))
print(len(sample_df))
```

```python
#correlation between variables
corr_df = df.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson", numeric_only=True)

plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
#regression line between loudness and energy
plt.figure(figsize=(11,7))
sns.regplot(data=sample_df, y="loudness", x="energy",color="b").set(title = "Energy vs Loudness")
```

```python
#regression line between popularity and acousticness
plt.figure(figsize=(11,7))
sns.regplot(data=sample_df, y="popularity", x="acousticness",color="c").set(title = "Popularity vs Acousticness")
```

```python
#top five artists 
top_five_artists = df.groupby("artists").count().sort_values(by="name",ascending=False)["name"][:5]
```

```python
#Artist with most danceable songs
top_five_artists_danceable_song=df[["danceability","name","artists"]].sort_values(by="danceability",ascending=False)[:5]
```

```python
plt.figure(figsize=(11,6))
sns.barplot(x="danceability",y="artists",data=top_five_artists_danceable_song, palette=['red','blue','green','orange','yellow'])
plt.title("Artists with the most danceable songs")
plt.show()
```
