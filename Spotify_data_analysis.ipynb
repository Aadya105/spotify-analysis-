```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("tracks.csv")

```

```python
# Save track_id separately (so we can use it later in output)
track_ids = df['id']

# Define target
y = df['popularity']

# Drop non-numeric columns including track_id
X = df.select_dtypes(include=['number'])

# Train-test split
x_train, x_test, y_train, y_test, id_train, id_test = train_test_split(
    X, y, track_ids, test_size=0.2, random_state=42
)

```

```python
# Train linear regression model
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
print("Training R²:", linear_model.score(x_train, y_train))
print("Testing R²:", linear_model.score(x_test, y_test))

# Predict on test set
linear_predict = linear_model.predict(x_test)

# Create output DataFrame with correct track IDs
l_output = pd.DataFrame({
    'track_id': id_test.values,
    'Predicted Popularity': linear_predict,
    'Actual Popularity': y_test.values
})

print(l_output.head(10))
```
